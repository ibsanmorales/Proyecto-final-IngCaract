```python
#librerias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Carga de datos
Utilizaremos los archivo generados en la libreta make_tidy_data.ipynb donde hicimos el procesamiento, un parte muy importante de ingenierias de caracteristicas obtencion, limpieza transformacion etc.

Los archivos con los que estremos trabajando son:
- educacion_tidy.csv
- presupuesto_tidy.csv

```python
path_edu = './tidy_data/educacion_tidy.csv'
path_presu = './tidy_data/presupuesto_tidy.csv'

df_edu = pd.read_csv(path_edu)

df_presu = pd.read_csv(path_presu)
```

```python
df_edu['Fecha'] = df_edu['Fecha'].astype('category')
df_presu['CICLO'] = df_presu['CICLO'].astype('category')
```

### Datos de Educacion

```python
df_edu.describe()
```

Generamos graficas para tener un mejor interpretacion de los datos

```python
#Grafica taza total de alfabetismo por año
fig, ax = plt.subplots(figsize = (12, 6))

g= sns.lineplot(x="Fecha", y="Alfabetización, total de adultos (% personas 15 o más)", data=df_edu, marker = 'o', alpha = 0.9,  palette="Blues_d")
g.axes.set_ylim(80,100)
plt.title('Taza de Alfabetizacion total por año (2008 - 2018)')
plt.xticks(rotation =0)
plt.xlabel('Fecha')
```

Vemos en la grafica superior que hay un reduccion en la taza en analfabetismo por año, parecindo un relazion positiva conforme a los años

```python
boxplot = sns.boxplot(
    y=df_edu['Alfabetización, total de adultos (% personas 15 o más)'])
```

```python
df_edu.dtypes
```

```python
fig, ax = plt.subplots(figsize = (12, 6))

ax= sns.lineplot(x='Fecha', y='licenciatura, población más de 25 años, total (%)', data=df_edu,marker = 'o')
ax = sns.lineplot(x='Fecha', y='maestría, población más de 25 años, total (%)', data=df_edu,marker = 'o')
ax = sns.lineplot(x='Fecha', y='doctorado, población más de 25 años, total (%)', data=df_edu, marker = 'o')
ax.legend(labels=['licenciatura', 'maestría','doctorado'])
ax.axes.set_ylim(0,20)
plt.ylabel("(% Porcentaje)")
plt.title('Taza de poblacion con Licenciatura | Maestria | Doctorado')
plt.xticks(rotation =0)
plt.xlabel('Fecha')
```

Se observa un claro cresimiento en la taza de poblacion con licenciatura comparado con la poblacion con maestria o doctorado se ve un tendencia mas estable o continua en su crecimiento

```python
boxplot = sns.boxplot(
    y=df_edu['maestría, población más de 25 años, total (%)'])
```

Se detecto un posible outlier al momento de realizar el boxplot, esto por curiosidad de pico que se tiene en el 2015 para la reprecentacion en gradica de linea de taza o promedio de poblacion con maestria.

```python
boxplot = sns.boxplot(
    y=df_edu['licenciatura, población más de 25 años, total (%)'])
```

### Datos de Presupuesto

```python
df_presu.describe()
```

Vemos que la descripcion de los datos los recursos ejercidos una anomalia muy grabe en el punto donde se tiene cero gasto ejercido, por ser portal de transparencia deberia de corroborrar si se asignaron recursos por que no se han subido o actualizado los montos ejercidos. 

```python
df_presu.dtypes
```

```python
#Grafica taza total de alfabetismo por año
fig, ax = plt.subplots(figsize = (16, 10))
ax = sns.barplot(x="CICLO", y="MONTO_APROBADO", data=df_presu, alpha = 1,  hue="DESC_TIPOGASTO",  palette="Paired", capsize=1, dodge=False,)

plt.title('Presupuesto anual Educacion')
plt.xticks(rotation =0)
plt.xlabel('Fecha')
```

Vemos por la grafica que el rubro predominante es el gasto corriente y conforme vamos avanzando en las fechas se va disminuyendo otros sectores

```python
fig, ax = plt.subplots(figsize = (12, 8))
ax =sns.scatterplot(x="CICLO", y="MONTO_APROBADO", hue="DESC_TIPOGASTO" ,data=df_presu,
             palette="deep")
ax.legend(labels=['Gasto corriente', 'Capita diferente de Obra p.','Obra publica','Corriente indirecto de subsidios', 'Inversion indirecto subsidio','Corriente recursos fideicomisos'])
```

En la utilma grafica observamos los montos por separado se tiene mejor visualizacion y compreson de ellos, como son montos puntuales un boxplot mos podria confundir y pensar que los puntos superirores son outlier, si tomamos el conjunto de datos como uno solo.
