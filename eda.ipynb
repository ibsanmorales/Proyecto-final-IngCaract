```python
#librerias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Carga de datos
Utilizaremos los archivo generados en la libreta make_tidy_data.ipynb donde hicimos el procesamiento, un parte muy importante de ingenierias de caracteristicas obtencion, limpieza transformacion etc.

Los archivos con los que estremos trabajando son:
- educacion_tidy.csv
- presupuesto_tidy.csv
- presupuesto_educacion.csv

```python
path_presu = './tidy_data/presupuesto_tidy.csv'
path_presu_edu = './tidy_data/presupuesto_educacion_tidy.csv'

df_presu = pd.read_csv(path_presu)
df_presu_edu = pd.read_csv(path_presu_edu)
```

```python
df_presu['CICLO'] = df_presu['CICLO'].astype('category')
df_presu_edu['CICLO'] = df_presu_edu['CICLO'].astype('category')
```

### Datos de Presupuesto Educacion

```python
df_presu_edu.describe()
```

Generamos graficas para tener un mejor interpretacion de los datos

```python
#Grafica taza total de alfabetismo por año
fig, ax = plt.subplots(figsize = (12, 6))

g= sns.lineplot(x="CICLO", y="Alfabetización, total de adultos (% personas 15 o más)", data=df_presu_edu, marker = 'o',  palette="Blues_d")
g.axes.set_ylim(80,100)
plt.title('Taza de Alfabetizacion total por año (2008 - 2018)')
plt.xlabel('CICLO')
```

Vemos en la grafica superior que hay un reduccion en la taza en analfabetismo por año, parecindo un relazion positiva conforme a los años

```python
fig, ax = plt.subplots(figsize = (12, 6))

ax= sns.lineplot(x='CICLO', y='licenciatura, población más de 25 años, total (%)', data=df_presu_edu,marker = 'o')
ax = sns.lineplot(x='CICLO', y='maestría, población más de 25 años, total (%)', data=df_presu_edu,marker = 'o')
ax = sns.lineplot(x='CICLO', y='doctorado, población más de 25 años, total (%)', data=df_presu_edu, marker = 'o')
ax.legend(labels=['licenciatura', 'maestría','doctorado'])
ax.axes.set_ylim(0,20)
plt.ylabel("(% Porcentaje)")
plt.title('Taza de poblacion con Licenciatura | Maestria | Doctorado')
plt.xticks(rotation =0)
plt.xlabel('CICLO')
```

```python
#Grafica taza total de alfabetismo por año
fig, ax = plt.subplots(figsize = (16, 10))
ax = sns.barplot(x="CICLO", y="MONTO_APROBADO", data=df_presu, alpha = 1,  hue="DESC_TIPOGASTO",  palette="Paired", capsize=1, dodge=False,)

plt.title('Presupuesto anual Educacion')
plt.xticks(rotation =0)
plt.xlabel('CICLO')
```

Vemos por la grafica que el rubro predominante es el gasto corriente y conforme vamos avanzando en las fechas se va disminuyendo otros sectores

```python
fig, ax = plt.subplots(figsize = (12, 8))
ax =sns.scatterplot(x="CICLO", y="MONTO_APROBADO", hue="DESC_TIPOGASTO" ,data=df_presu,
             palette="deep")
ax.legend(labels=['Gasto corriente', 'Capita diferente de Obra p.','Obra publica','Corriente indirecto de subsidios', 'Inversion indirecto subsidio','Corriente recursos fideicomisos'])
```

En la utilma grafica observamos los montos por separado se tiene mejor visualizacion y compreson de ellos, como son montos puntuales un boxplot mos podria confundir y pensar que los puntos superirores son outlier, si tomamos el conjunto de datos como uno solo.

Vamos a enfocarnos en todos los gasto excluyendo el gasto corriente

```python
df_presu_sincorriente = df_presu[df_presu['DESC_TIPOGASTO'] != 'Gasto corriente']
fig, ax = plt.subplots(figsize = (22, 10))
ax =sns.barplot(x="CICLO", y="MONTO_APROBADO", hue="DESC_TIPOGASTO",data=df_presu_sincorriente,
             alpha = 1,  palette="Paired", dodge=True)
plt.title('Presupuesto anual Educacion sin gasto corriente')
plt.xticks(rotation =0)
plt.xlabel('CICLO')
```

Observamos que los gastos adicionales al corriente se han reducido conforme el tiempo, manteninedo al final de los ultimos años presupuesto para obra publica mas estable pero menos a años anteriores (años < 2017).
